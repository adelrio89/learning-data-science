```python
import random

class Deck(object):
    
    def __init__(self):    
        self.deck = range(1,14) * 4
        
    def getCard(self):
        i = random.randint(0,len(self.deck)-1)
        print(len(self.deck))
        return self.deck.pop(i)
    
    def deal(self):
        card1 = self.getCard()
        card2 = self.getCard()
        return (card1, card2)
    
    def __len__(self):
        return len(self.deck)
            
    
class Player(object):
    
    def __init__(self, card1, card2, bet):
        self.hand = [card1, card2]
        self.bet = bet

    def hit(self, card):
        self.hand.append(card)
    
    def print_game(self, label):
        print("")
        print("GAME - %s:" %label)
        for i in range(len(self.hand)):
            if (i == 1 and label == 'DEALER'):
                continue
            else:
                ind = i + 1
                card = str(self.hand[i])
                print("Card %s: %s" %(ind,card))
    
    def getHand(self):
        tot = 0
        for i in self.hand:
            tot = tot + i
        return tot
            
  

    
def play():
   
    play = True
    deck = 52
    while play == True and deck > 0:
        
        print('WELCOME TO BLACKJACK!')
        d = Deck()
        d_card1, d_card2 = d.deal()
        p_card1, p_card2 = d.deal()
        bet = float(raw_input("How much would you like your bet to be?"))
        dealer = Player(d_card1,d_card2,0)
        player1 = Player(p_card1, p_card2, bet)
        dealer.print_game('DEALER')
        player1.print_game('PLAYER')
        pgame = player1.getHand()
        dgame = dealer.getHand()
        
        ##Blackjack
        if pgame == 21 and dgame <> 21:
            print("Player wins with a Blackjack!!")
            break

        else:
            win = ' '
            while not pgame > 21 and win == ' ':
                hit = raw_input("Would you like to hit or stand? h/s").lower().startswith('h')
                if hit == True:
                    new_card = d.getCard()
                    player1.hit(new_card)
                    dealer.print_game('DEALER')
                    player1.print_game('PLAYER')
                    pgame = player1.getHand()
                    dgame = dealer.getHand()
                    if pgame == 21 and dgame <> 21:
                        print("Player wins with a Blackjack")
                        win = 'Player'
                    elif pgame < 21 and dgame > 21:
                        print("Player wins with a hand of %s" %pgame)
                        win = 'Player'
                    elif pgame <> 21 and dgame == 21:
                        print("Dealer wins with a Blackjack")
                        win = 'Dealer'
                    elif pgame > 21 and dgame <= 21:
                        print("Player losts with a hand of %s" %pgame)
                        win = 'Dealer'
                    deck = len(d)
                else:
                    new_card = d.getCard()
                    dealer.hit(new_card)
                    dealer.print_game('DEALER')
                    player1.print_game('PLAYER')
                    pgame = player1.getHand()
                    dgame = dealer.getHand()
                    if pgame == 21 and dgame <> 21:
                        print("Player wins with a Blackjack")
                        win = 'Player'
                    elif pgame < 21 and dgame > 21:
                        print("Player wins with a hand of %s" %pgame)
                        win = 'Player'
                    elif pgame <> 21 and dgame == 21:
                        print("Dealer wins with a Blackjack")
                        win = 'Dealer'
                    elif pgame < dgame < 21:
                        print("Dealer wins with a hand of %s" %dgame)
                        win = 'Dealer'
                    elif pgame > 21 and dgame <= 21:
                        print("Player losts with a hand of %s" %pgame)
                        win = 'Dealer'
                    deck = len(d)
            else:
                play = raw_input("Want to play again? y/n").lower().startswith('y')
                deck = 52
    else:
        print("GAME OVER")
            
play()
```
